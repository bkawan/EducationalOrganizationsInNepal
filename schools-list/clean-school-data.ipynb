# Making copy of original DataFrame with necessary columns 

```python
school_data_df_copy = pd.DataFrame(data=school_data,columns=['School_name','District_name','Location','VDC/Municipality_name','Ward No.','Unnamed: 125'])
```

#### We can see first two rows are  NaN Rows and we need to drop them

```python
school_data_df_copy.head()
```

```python
school_data_df_copy.tail()
```

### Basic information of DataFrame
- We can see total rows 36239
- Total School_name rows equals to 36237
- some of them don't have ward number location 
- we need to fill Nan with some values
- For Location  we will fill NaN with "VDC/Municipality_name" 
- For Ward No. We will Fill NaN with 0

```python
school_data_df_copy.info()
```

```python
## droping first two rows
school_data_df1 = school_data_df_copy.drop(school_data_df_copy.index[[0,1]])

school_data_df1.head()
```

#### Basic info of New DataFrame

```python
school_data_df1.info()
```

```python
##filling Nan with values
school_data_df1['Location'] = school_data_df1['Location'].fillna(school_data_df1['VDC/Municipality_name'])
school_data_df1['Ward No.'] = school_data_df1['Ward No.'].fillna(0)
```

```python
school_data_df1.head()
```

```python
school_data_df1.info()
```

##### converting ward No and Unnamed:125 data type to int

```python
school_data_df1[['Ward No.','Unnamed: 125']] = school_data_df1[['Ward No.','Unnamed: 125']].astype(int)
```

```python
school_data_df2 = school_data_df1.rename(columns={"Unnamed: 125":"Total Sutdents: 2070"})
```

```python
school_data_df2.head()
```

```python
school_data_df2.info()
```

# Savinng Data Frame  to CSV
